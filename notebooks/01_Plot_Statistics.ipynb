# askEarth Explore - Plot data

This notebook describes how to plot statistics data that you downloaded from [askEarth Explore](https://i.ask.earth).

Make sure to first go through the instructions in [00_Getting_Started](./00_Getting_Started.ipynb).

```python
import pandas as pd

# STEP 1: On the askEarth Explore platform, after you made a search, you can download the statistics in the "Statistics" tab.

# STEP 2: Copy-paste the downloaded JSON here as data = ...
data = {"dates":["2023-08-10","2023-08-15"],"CAST_SHADOWS":[0,611],"VEGETATION":[439823,563074],"NOT_VEGETATED":[646647,519561],"UNCLASSIFIED":[93,3986],"THIN_CIRRUS":[669,0]}

# STEP 3: Plot the statistics with a pandas DataFrame and matplotlib
df = pd.DataFrame.from_dict(data)
df.plot(title="NDVI per SC2A classification", x="dates", xlabel="date", ylabel="# of pixels")
```
